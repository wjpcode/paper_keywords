```python
import numpy as np
import pandas as pd
import csv
```

```python
df2018=pd.read_csv('2018nn_kw.csv')
df2017=pd.read_csv('2017nn_kw.csv')
df2016=pd.read_csv('2016nn_kw.csv')
```

```python
#添加列索引名称
df2018.columns=['kw','num_kw','No.','info']
df2018[:5]
```

```python
df2017.columns=['kw','num_kw','No.','info']
df2017[:5]
```

```python
df2016.columns=['kw','num_kw','No.','info']
df2016[:5]
```

```python
#2018、2017即df_1817两年的数据进行合并
df_1817=pd.merge(df2018,df2017,how='outer',on='kw')
df_1817[:5]
```

```python
#df_1817再和16年的数据合并，得到df_181716三年的数据和
df_181716=pd.merge(df_1817,df2016,how='outer',on='kw')
df_181716[:5]
```

```python
#将索引命名为new_index
df_181716.index.name='new_index'
df_181716[:5]
```

```python
#对总表的索引列进行重命名
df_181716.columns=['kw','num_kw2018','No._2018','info_2018','num_kw2017','No._2017','info_2017',
                   'num_kw2016','No._2016','info_2016']
df_181716[:5]
```

```python
#数据透视化，移除非数值NaN，即将NaN转化为0
df_value=pd.pivot_table(df_181716,index=['new_index','kw'],values=['num_kw2018','num_kw2017','num_kw2016'],fill_value=0)
df_value[:5]
```

```python
#数据求和
df_value['sum_num_kw']=df_value['num_kw2016'] + df_value['num_kw2017'] + df_value['num_kw2018']
df_value[:5]
```

```python
#降序排列
df_final=df_value.sort_values('sum_num_kw',ascending=False)
df_final[:100]
```

```python
#保存数据
df_final.to_csv('nn_sum.csv')
```
